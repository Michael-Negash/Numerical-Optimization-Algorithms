# 5.) Implement the method of Lagrange multipliers

```python
import numpy as np
from scipy.optimize import fsolve, minimize
```

```python

def lagrangian_grad(z):
    # Unpack the variables and multipliers
    x1, x2, x3, x4, x5, l1, l2, l3 = z

    # Define f and its partial derivatives.
    # Note: for clarity, here we write the symbolic structure; in practice,
    # you would compute these derivatives either by hand or using automatic differentiation.
    
    # Gradients of the objective f
    df_dx1 = 2*(x1 - 1) + 2*(x1 - x2)
    df_dx2 = -2*(x1 - x2) + 2*(x2 - x3)
    df_dx3 = -2*(x2 - x3) + 4*(x3 - x4)**3
    df_dx4 = -4*(x3 - x4)**3 + 4*(x4 - x5)**3
    df_dx5 = -4*(x4 - x5)**3

    # Constraint functions
    h1 = x1 + x2**2 + x3**3 - (3*np.sqrt(2) + 2)
    h2 = x2 - x2**3 + x4 - (2*np.sqrt(2) - 2)
    h3 = x1 * x5 - 2

    # Gradients of constraints:
    # For h1:
    dh1_dx1 = 1
    dh1_dx2 = 2*x2
    dh1_dx3 = 3*x3**2
    dh1_dx4 = 0
    dh1_dx5 = 0

    # For h2:
    dh2_dx1 = 0
    dh2_dx2 = 1 - 3*x2**2
    dh2_dx3 = 0
    dh2_dx4 = 1
    dh2_dx5 = 0

    # For h3:
    dh3_dx1 = x5
    dh3_dx2 = 0
    dh3_dx3 = 0
    dh3_dx4 = 0
    dh3_dx5 = x1

    # Form the gradient equations (stationarity conditions)
    F = np.zeros(8)
    F[0] = df_dx1 + l1*dh1_dx1 + l3*dh3_dx1
    F[1] = df_dx2 + l1*dh1_dx2 + l2*dh2_dx2
    F[2] = df_dx3 + l1*dh1_dx3
    F[3] = df_dx4 + l2*dh2_dx4
    F[4] = df_dx5 + l3*dh3_dx5
    # Constraint equations:
    F[5] = h1
    F[6] = h2
    F[7] = h3
    return F


```

```python
# Initial guess for the variables and multipliers
z0 = np.array([1, 1, 1, 1, 2, 1, 1, 1])
# Solve the system using fsolve (a variant of Newton's method)
full_solution = fsolve(lagrangian_grad, z0)
solution = full_solution[:5]  # Extract only the first five values
# Define the objective function f(x) for the library routine.
def f(x):
    return (x[0]-1)**2 + (x[0]-x[1])**2 + (x[1]-x[2])**2 + (x[2]-x[3])**4 + (x[3]-x[4])**4

cons = [{'type': 'eq', 'fun': lambda x: x[0] + x[1]**2 + x[2]**3 - (3*np.sqrt(2)+2)},
        {'type': 'eq', 'fun': lambda x: x[1] - x[1]**3 + x[3] - (2*np.sqrt(2)-2)},
        {'type': 'eq', 'fun': lambda x: x[0]*x[4] - 2}]

res = minimize(f, x0=[1, 1, 1, 1, 2], constraints=cons, method='SLSQP', options={'ftol': 1e-9})

print("Display results of my implementation:")
print("Implementation solution x:", solution)
print("\n----------------------------------------------------")
print("Display results of the library routines:")
print("Library solution x =", res.x)
```
