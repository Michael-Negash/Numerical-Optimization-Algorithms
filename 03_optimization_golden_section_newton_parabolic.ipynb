# 3. Write implementations for the Golden Section Search, Newton’s method, and Parabolic Interpolation 

```python
import numpy as np

def golden_section_search(f, a, b, tol=1e-5):
    golden_ratio = (1 + np.sqrt(5)) / 2
    while b - a > tol:
        c = b - (b - a) / golden_ratio
        d = a + (b - a) / golden_ratio
        if f(c) < f(d):
            b = d
        else:
            a = c
    return (a + b) / 2
```

```python
def newtons_method(f, df, d2f, x0, tol=1e-5, max_iter=100):
    x = x0
    for _ in range(max_iter):
        dx = df(x) / d2f(x)
        x = x - dx
        if abs(dx) < tol:
            break
    return x
```

```python
def parabolic_interpolation(f, a, b, c, tol=1e-5, max_iter=100, scale_factor=1):
    for i in range(max_iter):
        # Evaluate function values (scaled)
        fa, fb, fc = f(a) * scale_factor, f(b) * scale_factor, f(c) * scale_factor

        # Compute the numerator and denominator for the vertex formula.
        numerator = (b - a)**2 * (fb - fc) - (b - c)**2 * (fb - fa)
        denominator = 2 * ((b - a) * (fb - fc) - (b - c) * (fb - fa))
        
        if denominator == 0:
            return b  # Prevent division by zero; return the current best guess.
        
        # Calculate the vertex of the parabola (x_min)
        x_min = b - numerator / denominator
        
        # Check for convergence: if the change is less than tol.
        if abs(x_min - b) < tol:
            return x_min

        # Update points for next iteration.
        # We choose the new triple by replacing the point with the highest function value.
        # One common strategy is:
        if x_min < b:
            # Evaluate function at new candidate.
            if f(x_min) < f(b):
                c = b
                b = x_min
            else:
                a = x_min
        else:
            if f(x_min) < f(b):
                a = b
                b = x_min
            else:
                c = x_min

    return b  # Return the best point found after max_iter iterations
```

```python
# Define the functions
f1 = lambda x: x**4 - 14*x**3 + 60*x**2 - 70*x
f2 = lambda x: 0.5*x**2 - np.sin(x)
f3 = lambda x: x**2 + 4*np.cos(x)

# Derivatives for Newton's Method
df1 = lambda x: 4*x**3 - 42*x**2 + 120*x - 70
d2f1 = lambda x: 12*x**2 - 84*x + 120

df2 = lambda x: x - np.cos(x)
d2f2 = lambda x: np.sin(x)

df3 = lambda x: 2*x - 4*np.sin(x)
d2f3 = lambda x: 2 - 4*np.cos(x)
```

```python
# Golden Section Search
min_f1_golden = golden_section_search(f1, 0, 3)
min_f2_golden = golden_section_search(f2, 0, 3)
min_f3_golden = golden_section_search(f3, 0, 3)

# Newton's Method (Starting point: 1.0 is arbitrary within the interval)
min_f1_newton = newtons_method(f1, df1, d2f1, x0=1.0)
min_f2_newton = newtons_method(f2, df2, d2f2, x0=1.0)
min_f3_newton = newtons_method(f3, df3, d2f3, x0=1.0)

# Parabolic Interpolation (Choosing three initial points within the interval)
min_f1_parabolic = parabolic_interpolation(f1, 0, 1, 3)
min_f2_parabolic = parabolic_interpolation(f2, 0, 1, 3)
min_f3_parabolic = parabolic_interpolation(f3, 0, 1, 3)

# Print results
print("Golden Section Search:")
print(f"f1: {min_f1_golden}")
print(f"f2: {min_f2_golden}")
print(f"f3: {min_f3_golden}")

print("\nNewton's Method:")
print(f"f1: {min_f1_newton}")
print(f"f2: {min_f2_newton}")
print(f"f3: {min_f3_newton}")

print("\nParabolic Interpolation:")
print(f"f1: {min_f1_parabolic}")
print(f"f2: {min_f2_parabolic}")
print(f"f3: {min_f3_parabolic}")

from scipy.optimize import minimize_scalar
# Minimization using scipy's library
res1 = minimize_scalar(f1, bounds=(0, 3), method='bounded')  # Use 'bounded' instead of 'brent'
res2 = minimize_scalar(f2, bounds=(0, 3), method='bounded')
res3 = minimize_scalar(f3, bounds=(0, 3), method='bounded')

print("\n----------------------------------------------------")
print("Display results of the library routines:")
print(f"f1: {min_f1_parabolic}")
print(f"f2: {min_f2_parabolic}")
print(f"f3: {min_f3_parabolic}")
```
