```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plotting the feasible region
x = np.linspace(0, 2, 100)
y1 = (6 - 5*x)   # From 5x1 + x2 <= 6
y2 = (6 - 3*x)/4  # From 3x1 + 4x2 <= 6
y3 = (6 - 4*x)/3  # From 4x1 + 3x2 <= 6

plt.figure(figsize=(10, 8))
plt.plot(x, y1, label=r'$5x_1 + x_2 \leq 6$')  # lable for 5x1 + x2 <= 6
plt.plot(x, y2, label=r'$3x_1 + 4x_2 \leq 6$') # lable for 3x1 + 4x2 <= 6
plt.plot(x, y3, label=r'$4x_1 + 3x_2 \leq 6$') # lable for 4x1 + 3x2 <= 6

# Shading the feasible region
plt.fill_between(x, np.minimum(y1, np.minimum(y2, y3)), 0, where=(x >= 0), interpolate=True, alpha=0.3)

# Plot the vertices
vertices = np.array([[0, 0], [1.2, 0], [0, 1.5], [6/7, 6/7],[12/11, 6/11]])
plt.scatter(vertices[:, 0], vertices[:, 1], color='red', label='Vertices')

# Contour lines for the objective function
X, Y = np.meshgrid(np.linspace(0, 2, 200), np.linspace(0, 2, 200))
Z = -3*X - 2*Y
plt.contour(X, Y, Z, levels=10, cmap='plasma')

plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title('Feasible Region and Objective Function Contours')
plt.legend()
plt.grid(True)
plt.show()
```
